```python
#importing libraries

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn

%matplotlib inline
```

```python
#loading the dataset
Bitcoin = pd.read_csv("BitcoinPrices.csv")
```

```python
Bitcoin.head()
```

```python
Bitcoin
```

```python
#summary statistics

Bitcoin.describe()
```

```python
#creating a new column series i.e return

Bitcoin['Return'] = Bitcoin['Adj Close'].pct_change()
```

```python
Bitcoin.head()
```

```python
#we calculate the average daily return of bitcoin according to our data
np.mean(Bitcoin['Return'])
```

```python
#plotting the closing prices

ax1 = Bitcoin['Adj Close'].plot(style = ['-'] )
```

```python
ax2 = Bitcoin['Return'].plot(style = ['-'])
```

```python
ax3 = Bitcoin['Return'].plot(style = 'k.')
```

```python
#studying some histograms

Bitcoin['Adj Close'].hist()
```

```python
Bitcoin['Return'].hist()
```

```python
#kernel density plots 

Bitcoin['Adj Close'].plot(kind = 'kde')
```

```python
Bitcoin['Return'].plot(kind = 'kde')
```

```python
#exponential smoothing (elementary time series analysis)

from statsmodels.tsa.holtwinters import SimpleExpSmoothing

model = SimpleExpSmoothing(Bitcoin['Adj Close']).fit(smoothing_level = .01 , optimized = False)

Bitcoin['Adj Close'].plot(marker = 'o', color = 'red')
model.fittedvalues.plot(marker = 'o', color = 'blue')
```

```python
#lag plot to check if the series is random

from pandas.plotting import lag_plot
lag_plot(Bitcoin['Adj Close'])
```

```python
#autocorrelation plot

from pandas.plotting import autocorrelation_plot

autocorrelation_plot(Bitcoin['Adj Close'])
```
